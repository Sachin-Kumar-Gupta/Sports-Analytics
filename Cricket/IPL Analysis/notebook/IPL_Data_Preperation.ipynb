```python
def get_extras_type(row):
    extras = []
    if row['wide'] > 0:
        extras.append('wide')
    if row['noballs'] > 0:
        extras.append('noball')
    if row['legbyes'] > 0:
        extras.append('legbye')
    if row['byes'] > 0:
        extras.append('bye')
    return ', '.join(extras) if extras else None

merged_df_2025['extras_type'] = merged_df_2025.apply(get_extras_type, axis=1)
```

```python
merged_df_2025 = merged_df_2025.drop(['wide', 'legbyes','byes', 'noballs'],axis = 1)
```

```python
def get_win_type(row):
    result = []
    if pd.notna(pd.to_numeric(row['wb_runs'], errors='coerce')) and pd.to_numeric(row['wb_runs'], errors='coerce') > 0:
        result.append('runs')
    if pd.notna(pd.to_numeric(row['wb_wickets'], errors='coerce')) and pd.to_numeric(row['wb_wickets'], errors='coerce') > 0:
        result.append('wickets')
    if (
        (isinstance(row['wb_wickets'], str) and 'super' in row['wb_wickets'].lower()) or
        (isinstance(row['wb_runs'], str) and 'super' in row['wb_runs'].lower())):
        result.append('super over')
    result = list(dict.fromkeys(result))
    return ', '.join(result) if result else None

merged_df_2025['result'] = merged_df_2025.apply(get_win_type, axis=1)
```

```python
merged_df_2025['result'].unique()
```

```python
def get_result_margin(row):
    # Convert to numeric if possible
    runs = pd.to_numeric(row['wb_runs'], errors='coerce')
    wkts = pd.to_numeric(row['wb_wickets'], errors='coerce')

    if pd.notna(runs) and runs > 0:
        return int(runs)
    elif pd.notna(wkts) and wkts > 0:
        return int(wkts)
    elif (
        isinstance(row['wb_runs'], str) and 'super' in row['wb_runs'].lower()
    ) or (
        isinstance(row['wb_wickets'], str) and 'super' in row['wb_wickets'].lower()
    ):
        return "super over"
    else:
        return None

merged_df_2025['result_margin'] = merged_df_2025.apply(get_result_margin, axis=1)
```

**🔍 Why Fuzzy Matching?**

Player names often appear in different formats (e.g., "Sai Sudharsan" vs "B Sai Sudharsan", "Rohit Sharma" vs "RG Sharma"), causing inconsistencies during analysis.

To handle such variations, we’ll use fuzzy matching to detect and map similar names accurately—saving time and improving data quality.

```python
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
unique_names = df_combined['striker'].dropna().unique()

# Set a threshold (say 85% similarity)
similar_names = []

for i, name in enumerate(unique_names):
    for other_name in unique_names[i+1:]:
        if fuzz.ratio(name, other_name) >= 80:
            similar_names.append((name, other_name, fuzz.ratio(name, other_name)))
```

```python
# View potential duplicates
similar_names_df = pd.DataFrame(similar_names, columns=['Name1', 'Name2', 'Similarity'])
display(similar_names_df.sort_values('Similarity', ascending=False))
```

```python
# For Bowler's
unique_bowler = df_combined['bowler'].dropna().unique()

# Set a threshold (say 85% similarity)
similar_names = []

for i, name in enumerate(unique_bowler):
    for other_name in unique_bowler[i+1:]:
        if fuzz.ratio(name, other_name) >= 90:
            similar_names.append((name, other_name, fuzz.ratio(name, other_name)))
```

```python
# View potential duplicates
similar_names_df = pd.DataFrame(similar_names, columns=['Name1', 'Name2', 'Similarity'])
display(similar_names_df.sort_values('Similarity', ascending=False))
```
